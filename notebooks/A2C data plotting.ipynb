```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def loadData(filename):
    f = open(filename,'r')
    data = []
    lines = f.readlines()
    for line in lines:
        data.append(float(line))
    f.close()
    return np.array(data)

def plotData(data,datalabel):
    xaxis = [i for i in range(len(data))]        
    plt.plot(xaxis,data,label=datalabel)
    plt.legend()
    plt.show()
```

```python
entropy = loadData("../Entropy")
plotData(entropy,"Entropy")
valueLoss = loadData("../ValueLoss")
plotData(valueLoss,"Value Loss")
actionGain = loadData("../ActionGain")
plotData(actionGain,"Policy Loss")
totalLoss = loadData("../TotalLoss")
plotData(totalLoss, "Total Loss")
reward = loadData("../Reward")
avReward = [0 for i in range(len(reward))]
avReward[0:1000] = [np.mean(reward[0:i+1]) for i in range(1000)]
avReward[1000:] = [np.mean(reward[i-999:i+1]) for i in range(1000,len(reward))]
plotData(avReward,"Average Reward")
```
